#pandas: parallel coordinate

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize'] = [6,5]
pd.plotting.parallel_coordinates(iris_df, 'target');
```

ดีสำหรับ data น้อยๆ

#pandas: scatter matrix

```python
pd.plotting.scatter_matrix(iris_df, figsize=(7,7));
```

ใน Scatter Matrix Plot เส้นทแยงมุมจะเป็นฮิสโตแกรม (แท่ง) เพื่อแสดงการกระจายตัวของข้อมูลสำหรับแต่ละตัวแปรของตัวมันเอง เหมือนตาราง correlation

```python
matplotlib.rcParams['figure.figsize'] = [15,45] #กำหนดขนาดของ figure

plt.imshow(iris_df.iloc[:,:],interpolation='nearest', cmap='rainbow')
```

```python
matplotlib.rcParams['figure.figsize'] = [6,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=['red','green','blue']); #ชื่อสี
```

```python
matplotlib.rcParams['figure.figsize'] = [6,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=['c','m','y']); #ตัวย่อชื่อสี c m y k r g b w
```

สี: ฟ้าอมเขียว (Cyan), แดงอมม่วง (Magenta), เหลือง (Yellow), และดำ (Key)

```python
matplotlib.rcParams['figure.figsize'] = [6,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=[[1,0,0],[0.5,0.5,0.5],[0,1,1]]); #ค่าสี R G B
```

```python
matplotlib.rcParams['figure.figsize'] = [6,5]
pd.plotting.parallel_coordinates(iris_df, 'target',color=['#715BB3','#72F774','#FFA257']); #html color code
```
