#Quiz 11

plot กราฟ 3 อันในแนวทแยงมุม

1.boxplot

2.scatter

3.histrogram

````
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

# Boxplot on the first diagonal subplot
axes[0, 0].boxplot(stopandsearch_df['Latitude'].dropna())
axes[0, 0].set_title('Boxplot of Latitude')
axes[0, 0].set_ylabel('Latitude')


# Scatter plot on the second diagonal subplot
axes[1, 1].scatter(stopandsearch_df['Latitude'].dropna(), stopandsearch_df['Longitude'].dropna())
axes[1, 1].set_title('Scatter plot of Latitude vs Longitude')
axes[1, 1].set_xlabel('Latitude')
axes[1, 1].set_ylabel('Longitude')


# Histogram on the third diagonal subplot
axes[2, 2].hist(stopandsearch_df['Latitude'].dropna(), bins=20)
axes[2, 2].set_title('Histogram of Latitude')
axes[2, 2].set_xlabel('Latitude')
axes[2, 2].set_ylabel('Frequency')

# Hide the unused subplots
for i in range(3):
    for j in range(3):
        if i != j:
            axes[i, j].remove()

plt.tight_layout()
plt.show()
````
